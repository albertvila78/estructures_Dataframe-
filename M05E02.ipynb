<h1>Exercici 1</h1>

```python
#Ordenar el CSV

import pandas as pd


df = pd.read_csv('housing data.csv', header=None)


df.columns = ['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX', 'PTRATIO', 'B', 'LSTAT', 'MEDV']


print(df)
```

```python
print(df.columns)
```

```python
#Divdir en train i test

from sklearn.model_selection import train_test_split

train_df, test_df = train_test_split(df, test_size=0.2, random_state=42)

train_df.to_csv('train.csv', index=False, header=True)
test_df.to_csv('test.csv', index=False, header=True)
```

<h1>Exercici 2</h1>

```python
#Creació de columnes dummies

train_df = pd.get_dummies(train_df, columns=['CHAS'], prefix='CHAS')
test_df = pd.get_dummies(test_df, columns=['CHAS'], prefix='CHAS')
```

```python
print(train_df)
```

```python
#Estandarització de les columnes númeriques

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

train_df[["CRIM", "ZN", "INDUS", "NOX", "RM", "AGE", "DIS", "TAX", "PTRATIO", "B", "LSTAT", "MEDV"]] = scaler.fit_transform(train_df[["CRIM", "ZN", "INDUS", "NOX", "RM", "AGE", "DIS", "TAX", "PTRATIO", "B", "LSTAT", "MEDV"]])
test_df[["CRIM", "ZN", "INDUS", "NOX", "RM", "AGE", "DIS", "TAX", "PTRATIO", "B", "LSTAT", "MEDV"]] = scaler.transform(test_df[["CRIM", "ZN", "INDUS", "NOX", "RM", "AGE", "DIS", "TAX", "PTRATIO", "B", "LSTAT", "MEDV"]])
```

```python
#Polinomis per les variables númeriques que representan relaciones no lineales, RM" (promitj de habitacions por habitatge) i "LSTAT" (% de estatus inferior de la població).

from sklearn.preprocessing import PolynomialFeatures


poly = PolynomialFeatures(degree=2)


train_poly = poly.fit_transform(train_df[["RM", "LSTAT"]])
test_poly = poly.transform(test_df[["RM", "LSTAT"]])


train_data_poly = pd.DataFrame(train_poly, columns=poly.get_feature_names_out(["RM", "LSTAT"]))
test_data_poly = pd.DataFrame(test_poly, columns=poly.get_feature_names_out(["RM", "LSTAT"]))


train_data_final = pd.concat([train_df, train_data_poly], axis=1)
test_data_final = pd.concat([test_df, test_data_poly], axis=1)

```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

train_df = pd.read_csv('train.csv')
test_df = pd.read_csv('test.csv')

train_stats = train_df.describe()
test_stats = test_df.describe()

print("Resumen estadístic del conjunto de entrenament:")
print(train_stats)

print("Resum estadístic del conjunt de prova:")
print(test_stats)
```

```python
#Veure la distribucio de les tases de crim en les dues mostres

import matplotlib.pyplot as plt

plt.hist(train_df['CRIM'])
plt.xlabel('Per capita crime rate by town')
plt.ylabel('Frequencia')
plt.title('Historiograma de CRIM in Train')
plt.show()


plt.hist(test_df['CRIM'])
plt.xlabel('Per capita crime rate by town')
plt.ylabel('Frequenciay')
plt.title('Historiograma de CRIM  in Test')
plt.show()
```

```python
import seaborn as sns

# Gràfic de barres de la variable 'CHAS' en el conjunt  train i test per veure com es se distribueix la variable binaria que indica
#si un vivienda habitatge limita o no amb el riu Charles.

sns.countplot(x='CHAS', data=train_df)
plt.xlabel('Riu Charles')
plt.ylabel('Frequencia')
plt.title('Gràfic de barres CHAS Train')
plt.show()


sns.countplot(x='CHAS', data=test_df)
plt.xlabel('Riu Charles')
plt.ylabel('Frequencia')
plt.title('Gràfic de barres CHAS Test')
plt.show()
```

```python
# Gráfico de dispersión entre 'RM' y 'MEDV' en el conjunto de train i test per entendre com es relacionen
#les variables i la relació lineal entre elles

plt.scatter(train_df['RM'], train_df['MEDV'])
plt.xlabel('Average number of rooms per dwelling')
plt.ylabel('Median value of owner-occupied homes')
plt.title('Scatter plot of RM vs MEDV in Train')
plt.show()

plt.scatter(test_df['RM'], test_df['MEDV'])
plt.xlabel('Average number of rooms per dwelling')
plt.ylabel('Median value of owner-occupied homes')
plt.title('Scatter plot of RM vs MEDV in Test')
plt.show()
```

```python
# Gráfico de caixa de la variable 'LSTAT' en el conjunt de train i test per entendre la distribució i variabilitat
#en les dues mostres
plt.boxplot(train_df['LSTAT'])
plt.xlabel('% lower status of the population')
plt.title('Box plot of LSTAT in Train')
plt.show()

plt.boxplot(test_df['LSTAT'])
plt.xlabel('% lower status of the population')
plt.title('Box plot of LSTAT in Test')
plt.show()
```

```python
import seaborn as sns

# Matriu de correlació en el conjunto de train i test per entendre les relacions entre las variables 
#i verificar si hi ha variablescorrelacionades que puguin afectar el rendiment del model.
corr_train = train_df.corr()
sns.heatmap(corr_train, annot=True)
plt.title('Correlation Matrix in Train')
plt.show()

# Matriz de correlación en el conjunto de test
corr_test = test_df.corr()
sns.heatmap(corr_test, annot=True)
plt.title('Correlation Matrix in Test')
plt.show()
```
