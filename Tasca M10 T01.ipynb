```python
import requests
from bs4 import BeautifulSoup


url = "https://es.wikipedia.org/wiki/Python"

# Realitzar la sol·licitud HTTP
response = requests.get(url)

if response.status_code == 200:
    html = response.text
    soup = BeautifulSoup(html, "html.parser")

    # Extreure el títul de la pàgina
    title = soup.find("title").text
    print("Títul de la pàgina:", title)

    # Extreure les capçeleres de secció
    headers = soup.find_all(["h1", "h2", "h3", "h4", "h5", "h6"])
    print("Capçeleras de secció:")
    for header in headers:
        print(header.text.strip())
else:
    print("Error obtenint la pàgina:", response.status_code)
```

```python
import requests
from bs4 import BeautifulSoup

url = "https://es.wikipedia.org/wiki/Python"

response = requests.get(url)

if response.status_code == 200:
    html = response.text
    soup = BeautifulSoup(html, "html.parser")

    # Trobar la secció de l'índex de continguts
    index_section = soup.find("ul", class_="vector-toc-contents")

    if index_section:
        
        links = index_section.find_all("a", href=True)

        for link in links:
            
            article_title = link.get_text()
            
            
            print(article_title)
    else:
        print("No s'ha trobat la secció de l'index de continguts.")
else:
    print("Error a l'obtenir la pàgina principal:", response.status_code)
```

```python
response = requests.get(url)

if response.status_code == 200:
    html = response.text
    soup = BeautifulSoup(html, "html.parser")

  
    index_section = soup.find("ul", class_="vector-toc-contents")

    if index_section:
        
        historia_link = index_section.find("a", href="#Historia")

        if historia_link:
            # Obtenir l'enllaç a Historia
            historia_url = "https://es.wikipedia.org/wiki/Python#Historia"

            # Realitzar sol·licitud HTTP a la pàgina Historia
            historia_response = requests.get(historia_url)

            if historia_response.status_code == 200:
                historia_html = historia_response.text
                historia_soup = BeautifulSoup(historia_html, "html.parser")

                # Trobar el contingut de l'artícle Historia
                historia_content = historia_soup.find("div", class_="mw-parser-output")

                if historia_content:
                    
                    # Extreure text de l'article Historia
                    historia_text = historia_content.text
                    print(historia_text)
                else:
                    print("No s'ha trobat el contingut de l'artícle Historia.")
            else:
                print("Error a l'obtenir la pàgina Historia:", historia_response.status_code)
        else:
            print("No s'ha trobat l'enllaç a l'article Historia en l'index de conteninguts.")
    else:
        print("No s'ha trobat la secció del índex de continguts.")
else:
    print("Error a l'obtenir la pàgina principal:", response.status_code)

```

```python
from selenium.webdriver.common.by import By

url = "https://www.laliga.com/es-GB"

driver = Firefox()

driver.get(url)

title = driver.title
print("Títul de la pàgina:", title)

headers = driver.find_elements(By.CSS_SELECTOR, "h1, h2, h3, h4, h5, h6")
print("Capçaleras de secció:")
for header in headers:
    print(header.text.strip())

driver.quit()
```

```python
from selenium.webdriver.common.by import By

url = "https://www.laliga.com/es-GB"

driver = Firefox()

driver.get(url)

internal_links = driver.find_elements(By.CSS_SELECTOR, "a[href]")


for link in internal_links:
    href = link.get_attribute("href")
    print(href)

driver.quit()
```
