<h1>Exercici 2: Pre-processament de dades</h1>

```python
movies = pd.read_csv("https://raw.githubusercontent.com/IT-Academy-BCN/Data-Science/main/Pre-processing-data/movies.dat",
sep='::', encoding = 'ISO-8859-1', index_col=0, names=['TtulAny','Genere'], engine='python')
```

```python
movies.head()
```

```python
movies['Genere'] = movies['Genere'].str.split('|').str[0]

print(movies)
```

```python
movies.head()
```

<h1>Exercici 3</h1>

```python
#Aqui volia veure quins generes han sigut els mes predominants al llarg del temps, en numeros globals, entenen que aix son produccions nomes nord-americanes
#He agafat el grfic de la bibilioteca de Seaborn

import seaborn as sns
import matplotlib.pyplot as plt
sns.set_theme(style="whitegrid")

f, ax = plt.subplots(figsize=(6, 15))

generes = movies['Genere'].value_counts()

sns.barplot(x=generes.values, y=generes.index, label="Total", color="b")

ax.legend(ncol=2, loc="lower right", frameon=True)

ax.set(xlim=(0, generes.max()), ylabel="Generes", xlabel="Total")

sns.despine(left=True, bottom=True)

plt.show()
```

```python
#Aqu volia veure quina ha sigut l'evoluci en la cuantitat produccions al llarg dels anys. Seria interesant veure una serie mes llarga fins el dia d'avui.

import seaborn as sns

sns.set_theme(style="whitegrid")

movies['Any'] = movies['TtulAny'].str.extract(r'\((\d{4})\)').astype(int)

movies_year = movies['Any'].value_counts().sort_index().reset_index()

movies_year.columns = ['Any', 'Total']

sns.lineplot(data=movies_year, x='Any', y='Total')
```
